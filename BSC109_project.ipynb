```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px

url = 'https://raw.githubusercontent.com/TannerMcCarthy/finalProjectDatacsv/main/crime%20stats%202010-2020%20revised.csv'

dataset = pd.read_csv(url)

ydata = dataset.iloc[:,1:12].values

year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

crimeA = dataset.iloc[1,1:12].values
crimeB = dataset.iloc[2,1:12].values
crimeC = dataset.iloc[3,1:12].values
crimeD = dataset.iloc[4,1:12].values
crimeE = dataset.iloc[5,1:12].values
crimeF = dataset.iloc[6,1:12].values
crimeG = dataset.iloc[7,1:12].values
crimeH = dataset.iloc[8,1:12].values

plt.figure(figsize=(15, 15))
plt.plot(year, crimeA, label = 'Murder')
plt.plot(year, crimeB, label = 'Rape')
plt.plot(year, crimeC, label = 'Robbery')
plt.plot(year, crimeD, label = 'Felony Assault')
plt.plot(year, crimeE, label = 'Burglary')
plt.plot(year, crimeF, label = 'Grand Larceny')
plt.plot(year, crimeG, label = 'Grand Larceny (Vehicle')
#plt.plot(year, crimeH, label = 'Total')

plt.title('Crime Rate Vs Year')
plt.xlabel('Year')
plt.ylabel('Crime Rate')
plt.legend()
plt.show()
```
